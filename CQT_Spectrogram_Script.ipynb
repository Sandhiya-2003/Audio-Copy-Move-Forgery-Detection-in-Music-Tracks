```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
sr = 16000
hop_length = 512
n_bins = 84
bins_per_octave = 12
```

# ***Original***

***Classical***

```python
audio_folder = "/content/drive/MyDrive/preprocessed_wav_tracks/classical"
image_output_folder = "/content/drive/MyDrive/CQT_Spectrograms/original/images"
npy_output_folder = "/content/drive/MyDrive/CQT_Spectrograms/original/npy_files"
```

```python
for file in os.listdir(audio_folder):
    if file.endswith(".wav"):
        file_path = os.path.join(audio_folder, file)

        base_filename = file.split(".")[0]

        y, _ = librosa.load(file_path, sr=sr)

        cqt = librosa.cqt(y, sr=sr, hop_length=hop_length, n_bins=n_bins, bins_per_octave=bins_per_octave)
        cqt_magnitude = np.abs(cqt)

        power_spectrum = cqt_magnitude**2
        log_power_spectrum = librosa.power_to_db(power_spectrum, ref=np.max)

        npy_filename = os.path.join(npy_output_folder, f"{base_filename}.npy")
        np.save(npy_filename, log_power_spectrum)

        plt.figure(figsize=(10, 5))
        librosa.display.specshow(log_power_spectrum, sr=sr, hop_length=hop_length, x_axis="time", y_axis="cqt_hz")
        plt.colorbar(label="dB")
        plt.title("CQT Spectrogram")
        image_filename = os.path.join(image_output_folder, f"{base_filename}_spectrogram_image.png")
        plt.savefig(image_filename)
        plt.close()

        print(f"{base_filename}.npy & {base_filename}_spectrogram_image.png")
```

```python
for file in os.listdir(audio_folder):
    if file.endswith(".wav"):
        file_path = os.path.join(audio_folder, file)

        base_filename = file.split("_")[0]

        y, _ = librosa.load(file_path, sr=sr)


        cqt = librosa.cqt(y, sr=sr, hop_length=hop_length, n_bins=n_bins, bins_per_octave=bins_per_octave)
        cqt_magnitude = np.abs(cqt)

        power_spectrum = cqt_magnitude**2
        log_power_spectrum = librosa.power_to_db(power_spectrum, ref=np.max)

        npy_filename = os.path.join(npy_output_folder, f"{base_filename}_forged.npy")
        np.save(npy_filename, log_power_spectrum)

        plt.figure(figsize=(10, 5))
        librosa.display.specshow(log_power_spectrum, sr=sr, hop_length=hop_length, x_axis="time", y_axis="cqt_hz")
        plt.colorbar(label="dB")
        plt.title("CQT Spectrogram")
        image_filename = os.path.join(image_output_folder, f"{base_filename}_forged_spectrogram_image.png")
        plt.savefig(image_filename)
        plt.close()

        print(f"{base_filename}_forged.npy & {base_filename}_forged_spectrogram_image.png")
```
